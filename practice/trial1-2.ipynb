```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
from control import bode, tf
from control.matlab import logspace
```

```python

plt.figure(1)
plt.subplot(2,1,1)
#plt.ylim(-20, 40)
plt.yticks(np.arange(-60, 41, 20))
plt.xlim(1e-1, 1e3)
plt.subplot(2,1,2)
plt.ylim(-360, 0)
plt.yticks(np.arange(-360, 1, 90))
plt.xlim(1e-1, 1e3)
fig1, fig2 = plt.gcf().axes # subplotのハンドラを取得
S = tf([10, 1],[10, 1+101**0.5])
mag, phase, om = bode(S, plot = False)
plt.sca(fig1)
plt.semilogx(om, 20*np.log10(mag))
plt.sca(fig2)
plt.semilogx(om, phase*180/np.pi)

S_1 = tf([10, 1, 0],[10, 11,1])
mag, phase, om = bode(S_1, plot = False)
plt.sca(fig1)
plt.semilogx(om, 20*np.log10(mag))
plt.sca(fig2)
plt.semilogx(om, phase*180/np.pi)
plt.xticks([])
```

```python
kp_s = np.arange(0.01, 2, 0.1)
plt.figure(1)
plt.subplot(2,1,1)
plt.ylim(-60, 40)
plt.yticks(np.arange(-60, 41, 20))
plt.xlim(1e-1, 1e3)
plt.subplot(2,1,2)
plt.ylim(-360, 0)
plt.yticks(np.arange(-360, 1, 90))
plt.xlim(1e-1, 1e3)
fig1, fig2 = plt.gcf().axes # subplotのハンドラを取得
for kp in range(len(kp_s)):
    S = tf([0, 1],[1, kp])
    mag, phase, om = bode(S, plot = False)
    plt.sca(fig1)
    plt.semilogx(om, 20*np.log10(mag))
    plt.sca(fig2)
    plt.semilogx(om, phase*180/np.pi)
plt.sca(fig1)
plt.xticks([])
```
