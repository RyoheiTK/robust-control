```python
P = tf([0, 1],[1, 1 ,1])
C = tf([2, 10, 3],[1, 0])
Gyr = feedback(P*C, 1)
gain, phase, w = bode(Gyr, logspace(-2,2), Plot=False)
fig, ax = plt.subplots(2,1, figsize=(4, 3.5))
ax[0].semilogx(w, 20*np.log10(gain))
ax[1].semilogx(w, phase*180/np.pi)
ax[0].grid(which="both", ls=':')
ax[0].set_ylabel('Gain [dB]')
ax[1].grid(which="both", ls=':')
ax[1].set_xlabel('$¥omega$ [rad/s]')
ax[1].set_ylabel('Phase [deg]')
fig.tight_layout()
```

```python
g = 9.81 # 重力加速度[m/s^2]
l = 0.2 # アームの長さ[m]
m = 0.5 # アームの質量[kg]
c = 1.5e-2 # 粘性摩擦係数
J = 1.0e-2 # 慣性モーメント
```

```python
P = tf([0,0,1],[J, c, m*g*l])
```

```python
C = tf([1],[1]) + tf([0.1,0],[0, 1]) + tf([0, 2],[1, 0])
Gyr = feedback(P*C, 1)
y,t = step(Gyr, np.arange(0, 10, 0.01))
fig, ax = plt.subplots(figsize=(3, 2.3))
ax.plot(t,y)
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.grid(ls=':')
```
